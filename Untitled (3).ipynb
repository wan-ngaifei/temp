```python
# ==========================================
# Cell 3: 基于 Cook-Torrance 的对抗投影仪 Shader
# ==========================================

class CookTorranceProjectorShader(nn.Module):

    def __init__(self, device="cpu", blend_params=None):
        super().__init__()
        self.device = device
        self.blend_params = blend_params if blend_params else BlendParams(
            background_color=(0.0, 0.0, 0.0)
        )

    # =========================
    # Cook-Torrance BRDF（材质模型）
    # =========================
    def cook_torrance_brdf(self, n, v, l, albedo, roughness, metallic):

        # 半角向量（view 和 light 的中间方向）
        h = F.normalize(v + l, dim=-1)

        # 各种点积（几何关系）
        NdotL = torch.clamp(torch.sum(n * l, dim=-1, keepdim=True), 1e-5, 1.0)
        NdotV = torch.clamp(torch.sum(n * v, dim=-1, keepdim=True), 1e-5, 1.0)
        NdotH = torch.clamp(torch.sum(n * h, dim=-1, keepdim=True), 0.0, 1.0)
        VdotH = torch.clamp(torch.sum(v * h, dim=-1, keepdim=True), 0.0, 1.0)

        # =========================
        # D项：法线分布（GGX）
        # =========================
        alpha = roughness ** 2
        alpha_sq = alpha ** 2

        denom = (NdotH ** 2) * (alpha_sq - 1.0) + 1.0
        D = alpha_sq / (torch.pi * denom ** 2 + 1e-7)

        # =========================
        # F项：Fresnel（Schlick近似）
        # =========================
        F0 = torch.lerp(torch.full_like(albedo, 0.04), albedo, metallic)
        F_schlick = F0 + (1.0 - F0) * ((1.0 - VdotH) ** 5)

        # =========================
        # G项：几何遮挡（Smith）
        # =========================
        k = ((roughness + 1.0) ** 2) / 8.0

        gl = NdotL / (NdotL * (1.0 - k) + k + 1e-7)
        gv = NdotV / (NdotV * (1.0 - k) + k + 1e-7)
        G = gl * gv

        # =========================
        # specular（镜面反射）
        # =========================
        specular = (D * F_schlick * G) / (4.0 * NdotL * NdotV + 1e-7)

        # =========================
        # diffuse（漫反射）
        # =========================
        kd = (1.0 - F_schlick) * (1.0 - metallic)
        diffuse = kd * albedo / torch.pi

        # 最终 BRDF（已经乘了 N·L）
        return (diffuse + specular) * NdotL

    # =========================
    # 主渲染函数
    # =========================
    def forward(self, fragments, meshes, **kwargs):

        # -------- 输入 --------
        cameras = kwargs.get("cameras")
        lights = kwargs.get("lights")
        projector_camera = kwargs.get("projector_camera")
        projector_image = kwargs.get("projector_image")

        # =========================
        # 1️⃣ 获取几何信息
        # =========================
        verts = meshes.verts_packed()
        faces = meshes.faces_packed()
        vertex_normals = meshes.verts_normals_packed()

        faces_verts = verts[faces]
        faces_normals = vertex_normals[faces]

        # 插值到像素
        world_coords_all = interpolate_face_attributes(
            fragments.pix_to_face,
            fragments.bary_coords,
            faces_verts
        )

        normals_all = interpolate_face_attributes(
            fragments.pix_to_face,
            fragments.bary_coords,
            faces_normals
        )

        world_coords = world_coords_all[:, :, :, 0, :]
        normals = F.normalize(normals_all[:, :, :, 0, :], dim=-1)

        # ⚠️ 修正法线方向（你的模型是反的）
        normals = -normals

        # =========================
        # 2️⃣ 材质（albedo）
        # =========================
        if meshes.textures is not None:
            albedo = meshes.sample_textures(fragments)[:, :, :, 0, :]
        else:
            albedo = torch.ones_like(world_coords) * 0.8

        # =========================
        # 3️⃣ 视角方向
        # =========================
        cam_pos = cameras.get_camera_center().view(-1, 1, 1, 3)
        view_dir = F.normalize(cam_pos - world_coords, dim=-1)

        # =========================
        # 4️⃣ 太阳光
        # =========================
        sun_pos = lights.location.view(-1, 1, 1, 3)
        sun_light_dir = F.normalize(sun_pos - world_coords, dim=-1)

        roughness = kwargs.get("roughness", torch.tensor([0.4], device=world_coords.device)).view(-1,1,1,1)
        metallic = kwargs.get("metallic", torch.tensor([0.8], device=world_coords.device)).view(-1,1,1,1)

        sun_brdf = self.cook_torrance_brdf(
            normals, view_dir, sun_light_dir, albedo, roughness, metallic
        )

        sun_reflection = sun_brdf * lights.diffuse_color.view(-1,1,1,3)

        # =========================
        # 5️⃣ 投影仪：几何投影
        # =========================
        B, H, W, _ = world_coords.shape
        world_coords_flat = world_coords.view(B, H * W, 3)

        proj_pts = projector_camera.transform_points_screen(world_coords_flat)
        proj_pts = proj_pts.view(B, H, W, 3)

        uv = proj_pts[..., :2]

        # 转到 [-1,1]
        uv[..., 0] = (uv[..., 0] / input_width) * 2 - 1
        uv[..., 1] = (uv[..., 1] / input_height) * 2 - 1
        uv[..., 1] = -uv[..., 1]

        if projector_image.shape[0] != B:
            projector_image = projector_image.expand(B, -1, -1, -1)

        sampled_proj_color = F.grid_sample(
            projector_image,
            uv,
            mode='bilinear',
            padding_mode='zeros',
            align_corners=False
        ).permute(0,2,3,1)

        # 只保留前方点
        valid_mask = (proj_pts[..., 2] > 0.0).float().unsqueeze(-1)
        sampled_proj_color = sampled_proj_color * valid_mask

        # =========================
        # 6️⃣ 投影光照
        # =========================
        proj_pos = projector_camera.get_camera_center().view(-1,1,1,3)
        proj_light_dir = F.normalize(proj_pos - world_coords, dim=-1)

        proj_brdf = self.cook_torrance_brdf(
            normals, view_dir, proj_light_dir, albedo, roughness, metallic
        )

        # ⚠️ 注意：你原代码这里 NdotL 没定义（这是一个真实 bug）
        NdotL = torch.clamp(torch.sum(normals * proj_light_dir, dim=-1, keepdim=True), 0.0, 1.0)

        PROJ_INTENSITY = 10

        proj_reflection = proj_brdf * sampled_proj_color * (NdotL + 0.3) * PROJ_INTENSITY

        # =========================
        # 7️⃣ 环境光
        # =========================
        ambient_color = lights.ambient_color.view(-1,1,1,3)
        ambient_reflection = albedo * ambient_color

        # =========================
        # 8️⃣ 最终合成
        # =========================
        final_color = torch.clamp(
            ambient_reflection + sun_reflection + proj_reflection,
            0.0, 1.0
        ).unsqueeze(-2)

        images = hard_rgb_blend(final_color, fragments, self.blend_params)

        return images


print("✅ 核心物理 Shader 编译完成！")
```

```python
# ==========================================
# Cell 4: 搭建 Sim-to-Real 物理融合沙盘 (同轴正面打击版)
# ==========================================
import torch
import os
from pytorch3d.io import load_objs_as_meshes
from pytorch3d.renderer import (
    TexturesVertex, PointLights, PerspectiveCameras, 
    RasterizationSettings, MeshRasterizer, MeshRenderer
)
class CustomRenderer(nn.Module):
    def __init__(self, rasterizer, shader, cameras):
        super().__init__()
        self.rasterizer = rasterizer
        self.shader = shader
        self.cameras = cameras   # ⭐ 存下来

    def forward(self, meshes, **kwargs):
        cameras = kwargs.get("cameras", self.cameras)  # ⭐ fallback
        fragments = self.rasterizer(meshes, cameras=cameras)
        kwargs["cameras"] = cameras   # ⭐ 强制覆盖
        images = self.shader(fragments, meshes, **kwargs)
        return images
print("🚀 正在加载真理级 3D 模型...")
obj_filename = os.path.abspath('tango_ground_truth.obj')
mesh = load_objs_as_meshes([obj_filename], device=device)

# 注入材质
verts_shape = mesh.verts_packed().shape
vertex_colors = torch.zeros((1, verts_shape[0], 3), device=device)
vertex_colors[0, 0:8, :] = torch.tensor([0.75, 0.75, 0.78], device=device)
vertex_colors[0, 8:12, :] = torch.tensor([0.05, 0.05, 0.15], device=device)
mesh.textures = TexturesVertex(verts_features=vertex_colors)

# 1. 敌方真实相机 (SPEED+ 完美焦距换算)
real_R = torch.tensor([[[0.43226265321629254, 0.8890468430852787, -0.1508134922164176], 
                        [0.8215678823535075, -0.45722278946341743, -0.34054887384994426], 
                        [-0.3717192668124495, 0.023303038329618025, -0.9280526682710544]]], dtype=torch.float32, device=device)
real_T = torch.tensor([[0.121934640527, -0.002725499453, 6.537296389389]], dtype=torch.float32, device=device)

fx_px, fy_px = 3200.0 * (768.0 / 1920.0), 3200.0 * (512.0 / 1200.0)
cx_px, cy_px = 960.0 * (768.0 / 1920.0), 600.0 * (512.0 / 1200.0)

enemy_camera = PerspectiveCameras(
    device=device, R=real_R, T=real_T,
    focal_length=((fx_px, fy_px),),
    principal_point=((cx_px, cy_px),),
    in_ndc=False, image_size=((512, 768),)
)

# 🌟 2. 伴飞投影仪：直接借用敌方相机的机位，保证 100% 命中正面！
projector_camera = enemy_camera

# 计算光源的世界坐标，并关闭环境光 (Black Room)
# 🌟 神级修复：直接向敌方相机索要它在真实宇宙中的绝对坐标！
cam_world_pos = enemy_camera.get_camera_center()

sun_light = PointLights(
    device=device, 
    location=cam_world_pos,  # 👈 灯泡现在100%死死绑在相机的正中心了！
    diffuse_color=[[0.0, 0.0, 0.0]], 
    ambient_color=[[0.0, 0.0, 0.0]]
)
input_width, input_height = 768, 512
raster_settings = RasterizationSettings(
    image_size=(input_height, input_width),
    blur_radius=0.0,
    faces_per_pixel=1,
    cull_backfaces=False   # ⭐ 关键开关
)
rasterizer = MeshRasterizer(
    cameras=enemy_camera,
    raster_settings=raster_settings
)
shader = CookTorranceProjectorShader(device=device)
renderer = CustomRenderer(
    rasterizer=rasterizer,
    shader=shader,
    cameras=enemy_camera   # ⭐ 必须
)
# === Step 2: 直接调用 rasterizer（不走 renderer）===
fragments = rasterizer(mesh)

pix_to_face = fragments.pix_to_face

visibility = (pix_to_face >= 0).float().mean()

print("可见像素比例:", visibility.item())
print("✅ 完美对齐沙盘布置完毕！正面同轴照射已锁定！")
print(renderer.rasterizer.cameras)
```

```python
# ==========================================
# Cell 5: 启动 Sim-to-Real 加性光照融合攻击
# ==========================================
import torchvision.transforms.functional as TF
from torchvision import transforms
from PIL import Image

input_width, input_height = cfg.DATASET.INPUT_SIZE
real_image_path = "/home/yifei/dataset/speedplus/speedplusv2/sunlamp/images/img000079.jpg"
real_pil_img = Image.open(real_image_path).convert("RGB").resize((input_width, input_height)) 
real_img_tensor = transforms.ToTensor()(real_pil_img).unsqueeze(0).to(device)

projector_image_tensor = torch.rand((1, 3, 512, 512), dtype=torch.float32, device=device) * 0.5 + 0.25 
projector_image = torch.nn.Parameter(projector_image_tensor)
optimizer = optim.Adam([projector_image], lr=0.05)

# 调大流明，确保光斑在真实环境光中足够刺眼
light_intensity = 2.5 
num_epochs = 20

print(f"💥 开始执行 Sim-to-Real 加性光照融合攻击，共 {num_epochs} 轮...")

for epoch in range(num_epochs):
    optimizer.zero_grad()
    rendered_mask = renderer(
        mesh, cameras=enemy_camera, lights=sun_light,
        projector_camera=projector_camera, projector_image=projector_image,
        roughness=torch.tensor([0.6], device=device), 
        metallic=torch.tensor([0.3], device=device)   
    )
    
    light_mask_tensor = rendered_mask[..., :3].permute(0, 3, 1, 2)
    # 物理加性融合
    spoofed_img = torch.clamp(real_img_tensor + light_mask_tensor * light_intensity, 0.0, 1.0)
    
    norm_image = TF.normalize(spoofed_img, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    
    outputs = pose_network(norm_image, gpu=device)
    heatmap_out = outputs[0]
    
    loss = -torch.mean(heatmap_out) 
    
    loss.backward()

    print("grad:", projector_image.grad.abs().mean().item())
    optimizer.step()
    
    with torch.no_grad():
        projector_image.clamp_(0.0, 1.0)
        
    if epoch % 10 == 0 or epoch == num_epochs - 1:
        cls_out, _, _, trans_out = outputs[1]
        best_idx = torch.argmax(cls_out[0])
        pred_trans = trans_out[0, best_idx, :]
        print(f"👉 Epoch {epoch:03d} | 幻觉激活强度: {-loss.item():.5f} | 敌方预测坐标: {pred_trans.detach().cpu().numpy()}")

final_view_to_plot = spoofed_img[0].detach().cpu().permute(1, 2, 0).numpy()
print("🎉 Sim-to-Real 攻击执行完毕！")
```
