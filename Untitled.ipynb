```python
# ==========================================
# Cell 1: 导入核心库与环境设定
# ==========================================
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.transforms.functional as TF
import numpy as np
import matplotlib.pyplot as plt

# 导入 PyTorch3D 的核心渲染组件
from pytorch3d.renderer import (
    RasterizationSettings, MeshRasterizer, MeshRenderer, 
    FoVPerspectiveCameras, PointLights, look_at_view_transform,
    BlendParams
)
from pytorch3d.renderer.blending import hard_rgb_blend
from pytorch3d.ops import interpolate_face_attributes
from pytorch3d.utils import ico_sphere
from pytorch3d.renderer import TexturesVertex

# 设定设备
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"🔥 当前运行设备: {device}")
```

```python
# ==========================================
# Cell 3: 基于 Cook-Torrance 的对抗投影仪 Shader
# ==========================================
class CookTorranceProjectorShader(nn.Module):
    def __init__(self, device="cpu", blend_params=None):
        super().__init__()
        self.device = device
        self.blend_params = blend_params if blend_params else BlendParams(background_color=(0.0, 0.0, 0.0))

    def cook_torrance_brdf(self, n, v, l, albedo, roughness, metallic):
        h = F.normalize(v + l, p=2, dim=-1)
        NdotL = torch.clamp(torch.sum(n * l, dim=-1, keepdim=True), min=1e-5, max=1.0)
        NdotV = torch.clamp(torch.sum(n * v, dim=-1, keepdim=True), min=1e-5, max=1.0)
        NdotH = torch.clamp(torch.sum(n * h, dim=-1, keepdim=True), min=0.0, max=1.0)
        VdotH = torch.clamp(torch.sum(v * h, dim=-1, keepdim=True), min=0.0, max=1.0)
        
        alpha = roughness ** 2
        alpha_sq = alpha ** 2
        denom = (NdotH ** 2) * (alpha_sq - 1.0) + 1.0
        D = alpha_sq / (torch.pi * denom ** 2 + 1e-7)
        
        F0 = torch.lerp(torch.full_like(albedo, 0.04), albedo, metallic)
        F_schlick = F0 + (1.0 - F0) * ((1.0 - VdotH) ** 5)
        
        k = ((roughness + 1.0) ** 2) / 8.0
        gl = NdotL / (NdotL * (1.0 - k) + k + 1e-7)
        gv = NdotV / (NdotV * (1.0 - k) + k + 1e-7)
        G = gl * gv
        
        specular = (D * F_schlick * G) / (4.0 * NdotL * NdotV + 1e-7)
        kd = (1.0 - F_schlick) * (1.0 - metallic)
        diffuse = kd * albedo / torch.pi
        
        return (diffuse + specular) * NdotL

    def forward(self, fragments, meshes, **kwargs):
        cameras = kwargs.get("cameras")                 
        lights = kwargs.get("lights")                   
        projector_camera = kwargs.get("projector_camera") 
        projector_image = kwargs.get("projector_image")   
        
        verts = meshes.verts_packed()
        faces = meshes.faces_packed()
        vertex_normals = meshes.verts_normals_packed()
        faces_verts = verts[faces]
        faces_normals = vertex_normals[faces]
        
        world_coords_all = interpolate_face_attributes(fragments.pix_to_face, fragments.bary_coords, faces_verts)
        normals_all = interpolate_face_attributes(fragments.pix_to_face, fragments.bary_coords, faces_normals)
        
        world_coords = world_coords_all[:, :, :, 0, :]
        normals = F.normalize(normals_all[:, :, :, 0, :], p=2, dim=-1)
        
        if meshes.textures is not None:
            albedo = meshes.sample_textures(fragments)[:, :, :, 0, :] 
        else:
            albedo = torch.ones_like(world_coords) * 0.8
            
        dev = world_coords.device
        roughness = kwargs.get("roughness", torch.tensor([0.4], device=dev)).view(-1, 1, 1, 1)
        metallic = kwargs.get("metallic", torch.tensor([0.8], device=dev)).view(-1, 1, 1, 1)
        
        cam_pos = cameras.get_camera_center().view(-1, 1, 1, 3)
        view_dir = F.normalize(cam_pos - world_coords, p=2, dim=-1)
        
        sun_pos = lights.location.view(-1, 1, 1, 3)
        sun_light_dir = F.normalize(sun_pos - world_coords, p=2, dim=-1)
        sun_brdf = self.cook_torrance_brdf(normals, view_dir, sun_light_dir, albedo, roughness, metallic)
        sun_reflection = sun_brdf * lights.diffuse_color.view(-1, 1, 1, 3)

        # 投影仪几何透视与物理反射
        B, H, W, _ = world_coords.shape
        world_coords_flat = world_coords.view(B, H * W, 3)
        proj_pts_flat = projector_camera.transform_points_ndc(world_coords_flat)
        proj_pts = proj_pts_flat.view(B, H, W, 3)
        uv_coords = proj_pts[..., :2]
        
        if projector_image.shape[0] != B:
            projector_image = projector_image.expand(B, -1, -1, -1)
            
        sampled_proj_color = F.grid_sample(
            projector_image, uv_coords, mode='bilinear', padding_mode='zeros', align_corners=False
        ).permute(0, 2, 3, 1) 
        
        valid_mask = (proj_pts[..., 2] > 0.0).float().unsqueeze(-1)
        sampled_proj_color = sampled_proj_color * valid_mask
        
        proj_pos = projector_camera.get_camera_center().view(-1, 1, 1, 3)
        proj_light_dir = F.normalize(proj_pos - world_coords, p=2, dim=-1)
        proj_brdf = self.cook_torrance_brdf(normals, view_dir, proj_light_dir, albedo, roughness, metallic)
        proj_reflection = proj_brdf * sampled_proj_color
        
        # 物理最终光照融合
        final_color = torch.clamp(sun_reflection + proj_reflection, 0.0, 1.0).unsqueeze(-2) 
        images = hard_rgb_blend(final_color, fragments, self.blend_params)
        return images
print("✅ 核心物理 Shader 编译完成！")
```

```python
# ==========================================
# Cell 4: 搭建真实物理攻防沙盘 (1:1 几何与物理材质还原)
# ==========================================
import torch
import os
from pytorch3d.io import load_objs_as_meshes
from pytorch3d.renderer import (
    TexturesVertex, PointLights, FoVPerspectiveCameras, 
    look_at_view_transform, RasterizationSettings, MeshRasterizer, MeshRenderer
)

print("🚀 正在加载真理级 3D 模型，并注入物理材质...")

# 1. 加载我们刚生成的终极模型
obj_filename = os.path.abspath('tango_ground_truth.obj')
mesh = load_objs_as_meshes([obj_filename], device=device)

# ==========================================
# 🌟 核心物理升级：分离机身与帆板的材质底色 (Albedo)
# ==========================================
# 获取顶点总数 (应该是 12 个)
verts_shape = mesh.verts_packed().shape
vertex_colors = torch.zeros((1, verts_shape[0], 3), device=device)

# 1. 机身材质 (顶点 0~7)：航空银灰色 (漫反射强，适合显示迷彩)
vertex_colors[0, 0:8, :] = torch.tensor([0.75, 0.75, 0.78], device=device) 

# 2. 太阳能帆板材质 (顶点 8~11)：深藏青/黑色 (吸收大量光线，仅保留高光)
vertex_colors[0, 8:12, :] = torch.tensor([0.05, 0.05, 0.15], device=device) 

# 将物理底色赋予模型
mesh.textures = TexturesVertex(verts_features=vertex_colors)
print("✅ 航天器肉体与物理隔热层/电池板加载完毕！")

# ==========================================
# 2. 布置光学与相机
# ==========================================
R_enemy, T_enemy = look_at_view_transform(dist=3.0, elev=15.0, azim=45.0)
enemy_camera = FoVPerspectiveCameras(device=device, R=R_enemy, T=T_enemy)

R_proj, T_proj = look_at_view_transform(dist=2.5, elev=0.0, azim=0.0)
projector_camera = FoVPerspectiveCameras(device=device, R=R_proj, T=T_proj)

# 环境太阳光与微弱的宇宙背景漫反射
sun_light = PointLights(
    device=device, 
    location=[[0.0, 2.0, 3.0]],      # 👈 挪到正面和相机的头顶！
    diffuse_color=[[1.5, 1.5, 1.5]], # 👈 亮度突破 1.0，模拟刺眼的太空太阳光！
    ambient_color=[[10.8, 10.8, 10.8]]  # 👈 环境底光拉高，确保暗部绝对可见！
)

# 3. 组装终极物理渲染器
input_width, input_height = cfg.DATASET.INPUT_SIZE
raster_settings = RasterizationSettings(
    image_size=(input_height, input_width), 
    blur_radius=0.0, 
    faces_per_pixel=1, 
)

renderer = MeshRenderer(
    rasterizer=MeshRasterizer(raster_settings=raster_settings),
    shader=CookTorranceProjectorShader(device=device) 
)
print("✅ 攻防沙盘布置完毕！")
```

```python
# ==========================================
# Cell 5: 启动热力图致盲攻击优化 (真理级物理材质版)
# ==========================================
# 1. 武器上膛：初始化投影幻灯片 [B, C, H, W]
projector_image = torch.ones((1, 3, 512, 512), device=device, requires_grad=True)
projector_image.data.fill_(1.0)  # 👈 初始为纯白色满功率光斑！
optimizer = optim.Adam([projector_image], lr=0.05)

num_epochs = 100
print(f"💥 开始执行核弹级 Heatmap 致盲攻击，共 {num_epochs} 轮...")

for epoch in range(num_epochs):
    optimizer.zero_grad()
    
    # A. 物理光线追踪渲染 (折中参数：模拟现实中高光帆板与哑光机身的综合光学效果)
    rendered_images = renderer(
        mesh, cameras=enemy_camera, lights=sun_light,
        projector_camera=projector_camera, projector_image=projector_image,
        roughness=torch.tensor([0.6], device=device), # 中等粗糙度，保证机身能显现迷彩
        metallic=torch.tensor([0.3], device=device)   # 中低金属度，避免陷入“黑镜效应”
    )
    
    # 🌟 修复点：把你刚才不小心删掉的这一行加回来！提取 RGB 通道并转换维度！
    rgb_image = rendered_images[..., :3].permute(0, 3, 1, 2)
    
    # B. ImageNet 归一化 (防域偏移)
    norm_image = TF.normalize(rgb_image, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    
    # C. 过 SPNv2 靶子
    outputs = pose_network(norm_image, gpu=device)
    
    # D. Heatmap 幻觉攻击
    heatmap_out = outputs[0]
    # 目标：最大化整个热力图的激活值，让网络产生满屏关键点的幻觉
    loss = -torch.mean(heatmap_out) 
    
    loss.backward()
    optimizer.step()
    
    # E. 物理截断 (光不能为负或过曝)
    with torch.no_grad():
        projector_image.clamp_(0.0, 1.0)
        
    if epoch % 10 == 0 or epoch == num_epochs - 1:
        # 顺便看看最终姿态回归崩溃成什么样了
        cls_out, _, _, trans_out = outputs[1]
        best_idx = torch.argmax(cls_out[0])
        pred_trans = trans_out[0, best_idx, :]
        print(f"👉 Epoch {epoch:03d} | 幻觉激活强度(上升中): {-loss.item():.5f} | 敌方预测坐标跳变: {pred_trans.detach().cpu().numpy()}")

# 保存最终渲染结果供下一步画图使用
final_view_to_plot = rgb_image[0].detach().cpu().permute(1, 2, 0).numpy()
print("🎉 攻击执行完毕！")
```

```python
# ==========================================
# Cell 6: 开盲盒与成果展示
# ==========================================
import numpy as np
import matplotlib.pyplot as plt

# 提取优化的光斑
optimized_pattern = projector_image[0].detach().cpu().permute(1, 2, 0).numpy()
optimized_pattern = np.clip(optimized_pattern, 0.0, 1.0)

# 提取敌方相机的最终视角
final_view = np.clip(final_view_to_plot, 0.0, 1.0)


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.imshow(optimized_pattern)
plt.title("Optimized Physical Camouflage (Projector 2D Pattern)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(final_view)
plt.title("What the Enemy Camera Sees (Spoofed View)")
plt.axis("off")

plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

print("🛠️ 正在进行微雕手术：剔除巨大三角翼，换装极细的 1cm 真实物理天线...")

# 1. 底部机身点 (Z=0)
bottom_pts = np.array([
    [-0.3700, -0.2640,  0.0000], # 1: 底 左后
    [-0.3700,  0.3040,  0.0000], # 2: 底 左前
    [ 0.3700,  0.3040,  0.0000], # 3: 底 右前
    [ 0.3700, -0.2640,  0.0000], # 4: 底 右后
])

# 2. 机身顶部点 (Z=0.3215)
body_top_pts = np.copy(bottom_pts)
body_top_pts[:, 2] = 0.3215       # 5, 6, 7, 8

# 3. 顶置超大太阳能帆板
panel_pts = np.array([
    [-0.3700, -0.3850,  0.3215], # 9:  帆 左后
    [-0.3700,  0.3850,  0.3215], # 10: 帆 左前
    [ 0.3700,  0.3850,  0.3215], # 11: 帆 右前
    [ 0.3700, -0.3850,  0.3215], # 12: 帆 右后
])

# 4. 提取原始天线尖端 (Ground Truth)，并制造 1cm 的微小偏移用于生成“细针面片”
ant1_tip = np.array([-0.5427,  0.4877,  0.2535])
ant2_tip = np.array([ 0.5427,  0.4877,  0.2591])
ant3_tip = np.array([ 0.3050, -0.5790,  0.2515])

# 制造宽度仅为 1cm 的辅助点
offset = np.array([0.01, 0.0, 0.0])

antenna_pts = np.array([
    ant1_tip, ant1_tip + offset, # 13, 14: 天线 1
    ant2_tip, ant2_tip + offset, # 15, 16: 天线 2
    ant3_tip, ant3_tip + offset, # 17, 18: 天线 3
])

verts = np.vstack((bottom_pts, body_top_pts, panel_pts, antenna_pts))

# 5. 组装拓扑面片 (法线朝外)
faces_obj = [
    # --- 长方体机身 ---
    [1, 4, 3], [1, 3, 2], # 底面
    [5, 6, 7], [5, 7, 8], # 顶面
    [1, 2, 6], [1, 6, 5], # 左面
    [4, 8, 7], [4, 7, 3], # 右面
    [2, 3, 7], [2, 7, 6], # 前面
    [1, 5, 8], [1, 8, 4], # 后面

    # --- 顶置超大帆板 ---
    [9, 10, 11], [9, 11, 12], 
    [9, 12, 11], [9, 11, 10], 

    # --- 极细微雕天线 (双面可见的 1cm 极窄长条) ---
    # 天线 1 连到机身左前边缘 V6
    [13, 6, 14], [13, 14, 6],
    # 天线 2 连到机身右前边缘 V7
    [15, 7, 16], [15, 16, 7],
    # 天线 3 连到机身右后边缘 V8
    [17, 8, 18], [17, 18, 8]
]

# 6. 写入 OBJ 文件
obj_path = os.path.abspath('tango_slim_antennas.obj')
with open(obj_path, 'w') as f:
    f.write("# Ultimate Tango (Body + Panel + Slim 1cm Antennas)\n")
    for v in verts:
        f.write(f"v {v[0]:.6f} {v[1]:.6f} {v[2]:.6f}\n")
    for face in faces_obj:
        f.write(f"f {face[0]} {face[1]} {face[2]}\n")

print(f"✅ 微雕天线换装成功！已保存至：{obj_path}")

# 可视化...
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
faces_idx = np.array(faces_obj) - 1
mesh_all = Poly3DCollection(verts[faces_idx], alpha=0.8, edgecolor='black', linewidths=0.5)
mesh_all.set_facecolor('silver')
ax.add_collection3d(mesh_all)

original_11_pts = np.vstack((panel_pts, bottom_pts, np.array([ant1_tip, ant2_tip, ant3_tip])))
ax.scatter(original_11_pts[:,0], original_11_pts[:,1], original_11_pts[:,2], color='red', s=40, zorder=5, label='11 GT Points')

mid_x, mid_y, mid_z = 0, 0, 0.16
max_range = 0.6 
ax.set_xlim(mid_x - max_range, mid_x + max_range)
ax.set_ylim(mid_y - max_range, mid_y + max_range)
ax.set_zlim(mid_z - max_range, mid_z + max_range)

ax.view_init(elev=20, azim=60)
plt.title("Tango with Realistic Slim Antennas")
plt.show()
```

```python
import scipy.io
import os
import numpy as np
import matplotlib.pyplot as plt

print("🛠️ 正在对 tangoPoints.mat 进行 X光扫描...")

# 1. 挂载路径
mat_path = os.path.abspath('spnv2/core/utils/models/tangoPoints.mat')

try:
    data = scipy.io.loadmat(mat_path)
    pts = None
    for key, val in data.items():
        if isinstance(val, np.ndarray) and val.shape in [(11, 3), (3, 11)]:
            pts = val if val.shape == (11, 3) else val.T
            break
            
    if pts is not None:
        print("\n🎯 提取到的 11 个 Ground Truth 锚点坐标：")
        for i, p in enumerate(pts):
            print(f"点 {i:02d}: [ {p[0]:8.4f}, {p[1]:8.4f}, {p[2]:8.4f} ]")
            
        # 2. 画出 3D 骨架并打上编号
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(pts[:,0], pts[:,1], pts[:,2], c='red', s=50)
        
        for i, p in enumerate(pts):
            ax.text(p[0], p[1], p[2], f' {i}', size=15, fontweight='bold', color='black')
            
        ax.set_xlabel('X (m)')
        ax.set_ylabel('Y (m)')
        ax.set_zlabel('Z (m)')
        plt.title("Tango 11 Keypoints Skeleton")
        plt.show()
        
    else:
        print("❌ 没找到 11x3 的坐标数据，请检查 mat 文件！")

except Exception as e:
    print(f"❌ 运行报错: {e}")
```
